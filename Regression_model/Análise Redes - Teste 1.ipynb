# Identificar as redes que utilizaram determinada feature e calcular as medias dos cos_sim e loss

```python
y = testes["features_col"]
y = y.str.replace('[','')
y = y.str.replace(']','')
y = y.str.replace('\r\n','')
```

```python
features = y.str.get_dummies(sep=' ')
features
```

```python
features_numbers = features.columns
loss = testes["test_loss"].values
cos_sim = testes["test_cos_sim"].values
media_loss = np.zeros(len(features_numbers))
media_cos_sim = np.zeros(len(features_numbers))
cont = 0
for i in features_numbers:
    #print(i)
    presence = features[i].values
    #print(np.sum(presence))
    media_loss[cont] = np.mean(loss[presence==1])
    media_cos_sim[cont] = np.mean(cos_sim[presence==1])
    cont = cont + 1
```

```python
dados = {"Feature":features_numbers, "Media loss":media_loss, "Media cos sim":media_cos_sim}
df = pd.DataFrame(dados)
df
```

```python
np.max(media_cos_sim)
```

```python
np.min(media_cos_sim)
```

Teste 1 obs:Todas as features tiveram media de loss e cos_sim parecidas. Verificou-se que todas as features apareciam em media 22 mil vezes para o total de 45 mil testes.
